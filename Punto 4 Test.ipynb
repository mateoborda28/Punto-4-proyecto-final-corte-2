```python
url = "http://127.0.0.1:8000/upload-imagen/"
file_path = "C:/Users/mbr19/OneDrive/Escritorio/Ultimo semestre/Python/Segundo Corte/PROYECTO FINAL IA CORTE 2-20241012/Punto 4/input_files/james1.png"

with open(file_path, "rb") as file:
    response = requests.post(url, files={"file": file})

if response.status_code == 200:
    image = Image.open(BytesIO(response.content))
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```
